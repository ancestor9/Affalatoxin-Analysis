```python
# prompt: 위경도가 포함된 주소정보를 100개만 생성해줘

import pandas as pd
import random

# 샘플 주소 데이터 (실제 주소가 아닌 임의의 형식입니다)
sample_streets = ["강남대로", "테헤란로", "세종대로", "을지로", "명동길"]
sample_districts = ["강남구", "서초구", "중구", "종로구", "마포구"]
sample_cities = ["서울특별시"]

data = []
for i in range(100):
    city = random.choice(sample_cities)
    district = random.choice(sample_districts)
    street = random.choice(sample_streets)
    building_number = random.randint(1, 200)
    address = f"{city} {district} {street} {building_number}"

    # 임의의 위경도 생성 (실제 위치와 일치하지 않음)
    latitude = random.uniform(37.4, 37.6)
    longitude = random.uniform(126.9, 127.1)

    data.append({
        '주소': address,
        '위도': latitude,
        '경도': longitude
    })

df = pd.DataFrame(data)
df
```

```python
# prompt: df에 추가컬럼으로 -14부터 40까지 임의의 수를 생성해줘

df['기온'] = [random.randint(-14, 40) for _ in range(len(df))]
df
```

```python
tf = df.sample(5)
tf.drop(columns=['기온'], inplace=True)
tf
```

```python
# prompt: tf의 데이터 위도경도와 가장 가까운 3개 행을 nearest neighbors 알고리즘으로 df에서 찾은 후에 3개 행의 '기온' 값의 평균을 tf의 새로운 '추정기온'으로 만들어줘.. 아주 간단하게 코드를 작성해줘

from sklearn.neighbors import NearestNeighbors
import numpy as np

# NearestNeighbors 모델 학습
nn = NearestNeighbors(n_neighbors=3)
nn.fit(df[['위도', '경도']])

# tf의 각 행에 대해 가장 가까운 3개 행 찾고 기온 평균 계산
estimated_temperatures = []
for index, row in tf.iterrows():
    distances, indices = nn.kneighbors([[row['위도'], row['경도']]])
    nearest_temps = df.iloc[indices[0]]['기온'].mean()
    estimated_temperatures.append(nearest_temps)

# tf에 추정기온 추가
tf['추정기온'] = estimated_temperatures

tf
```

```python
# prompt: /usr/local/lib/python3.11/dist-packages/sklearn/utils/validation.py:2739: UserWarning: X does not have valid feature names, but NearestNeighbors was fitted with feature names
#   warnings.warn(
# /usr/local/lib/python3.11/dist-packages/sklearn/utils/validation.py:2739: UserWarning: X does not have valid feature names, but NearestNeighbors was fitted with feature names
#   warnings.warn(
# /usr/local/lib/python3.11/dist-packages/sklearn/utils/validation.py:2739: UserWarning: X does not have valid feature names, but NearestNeighbors was fitted with feature names
#   warnings.warn(
# /usr/local/lib/python3.11/dist-packages/sklearn/utils/validation.py:2739: UserWarning: X does not have valid feature names, but NearestNeighbors was fitted with feature names
#   warnings.warn(
# /usr/local/lib/python3.11/dist-packages/sklearn/utils/validation.py:2739: UserWarning: X does not have valid feature names, but NearestNeighbors was fitted with feature names
#   warnings.warn( 를 해결해주는 코드를 다시 작성해줘

import pandas as pd
import random
from sklearn.neighbors import NearestNeighbors
import numpy as np


# 샘플 주소 데이터 (실제 주소가 아닌 임의의 형식입니다)
sample_streets = ["강남대로", "테헤란로", "세종대로", "을지로", "명동길"]
sample_districts = ["강남구", "서초구", "중구", "종로구", "마포구"]
sample_cities = ["서울특별시"]

data = []
for i in range(100):
    city = random.choice(sample_cities)
    district = random.choice(sample_districts)
    street = random.choice(sample_streets)
    building_number = random.randint(1, 200)
    address = f"{city} {district} {street} {building_number}"

    # 임의의 위경도 생성 (실제 위치와 일치하지 않음)
    latitude = random.uniform(37.4, 37.6)
    longitude = random.uniform(126.9, 127.1)

    data.append({
        '주소': address,
        '위도': latitude,
        '경도': longitude
    })

df = pd.DataFrame(data)
df

df['기온'] = [random.randint(-14, 40) for _ in range(len(df))]
df
tf = df.sample(5)
tf.drop(columns=['기온'], inplace=True)
tf


# NearestNeighbors 모델 학습
# X does not have valid feature names 경고를 피하기 위해 NumPy 배열로 변환
nn = NearestNeighbors(n_neighbors=3)
nn.fit(df[['위도', '경도']].values)

# tf의 각 행에 대해 가장 가까운 3개 행 찾고 기온 평균 계산
estimated_temperatures = []
# X does not have valid feature names 경고를 피하기 위해 NumPy 배열로 변환
tf_coords = tf[['위도', '경도']].values

# 모든 tf 좌표에 대해 한 번에 k neighbors를 찾습니다.
distances, indices = nn.kneighbors(tf_coords)

# 찾은 이웃 인덱스를 사용하여 tf의 각 행에 대한 추정 기온 계산
for row_indices in indices:
    nearest_temps = df.iloc[row_indices]['기온'].mean()
    estimated_temperatures.append(nearest_temps)


# tf에 추정기온 추가
tf['추정기온'] = estimated_temperatures

tf
```

```python
# prompt: 서울특별시 종로구 세종대로 58  와 위도경도가 가장 가까운 3개 해을 추출한 후 평균을 보여줘

# 대상 주소의 위경도 (임의의 값, 실제 위치와 일치하지 않음)
target_latitude = 37.575
target_longitude = 126.978

# NearestNeighbors 모델 학습 (전체 데이터 사용)
nn_full = NearestNeighbors(n_neighbors=3)
nn_full.fit(df[['위도', '경도']].values)

# 대상 위경도와 가장 가까운 3개 데이터 포인트 찾기
distances, indices = nn_full.kneighbors([[target_latitude, target_longitude]])

# 가장 가까운 3개 데이터 포인트의 기온 추출
nearest_temps = df.iloc[indices[0]]['기온']

# 기온 평균 계산
average_temp = nearest_temps.mean()

print("대상 주소와 가장 가까운 3개 지점의 기온:")
print(nearest_temps)
print(f"\n가장 가까운 3개 지점의 평균 기온: {average_temp:.2f}")
```
