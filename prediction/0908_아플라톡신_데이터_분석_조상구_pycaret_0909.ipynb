```python
# Save the modified DataFrame to a pickle file
import pandas as pd
data= pd.read_csv('/teamspace/studios/this_studio/data.csv')
display(data.head())
print(data.shape)
```

```python
# fix_imbalance = True ÌååÎùºÎØ∏ÌÑ∞Î•º Ï∂îÍ∞ÄÌïòÏó¨ SMOTE Ï†ÅÏö©
# Import the classification module
from pycaret.classification import *
s = setup(data, 
          target = 'JDGMNT_WORD_NAME_encoded', 
          session_id = 123,
          fix_imbalance = True,
           rare_to_value = 0.1) # Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Îì±Ïû• ÎπÑÏú®Ïù¥ 10% ÎØ∏ÎßåÏù∏ Ìù¨Í∑Ä Î≤îÏ£ºÎì§ÏùÑ ÌïòÎÇòÎ°ú Î¨∂Ïñ¥ ÏÉàÎ°úÏö¥ Î≤îÏ£ºÎ°ú Î≥ÄÌôò)
```

```python
# Î™®Îì† Î™®Îç∏ÏùÑ ÎπÑÍµêÌïòÏó¨ Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ï¢ãÏùÄ Î™®Îç∏ÏùÑ Ï∞æÏäµÎãàÎã§.
best_model = compare_models(fold=5)  # K-fold 5Î≤àÎßå, defaultÎäî 10Î≤àÏûÑ

# best_model Î≥ÄÏàòÏóê Í∞ÄÏû• Ï¢ãÏùÄ ÏÑ±Îä•ÏùÑ Î≥¥Ïù∏ Î™®Îç∏Ïù¥ Ï†ÄÏû•Îê©ÎãàÎã§.
print(best_model)
```

```python
# 1. ÏõêÌïòÎäî ÌäπÏ†ï Î™®Îç∏ ÏÉùÏÑ±
# Î™®Îç∏ IDÎäî PyCaretÏùò Î™®Îì† ÏßÄÏõê Î™®Îç∏ Î¶¨Ïä§Ìä∏ÏóêÏÑú ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
# LightGBM Î™®Îç∏ÏùÑ 3-Í≤π ÍµêÏ∞® Í≤ÄÏ¶ùÏúºÎ°ú ÏÉùÏÑ±Ìï©ÎãàÎã§.
lightgbm_model = create_model('lightgbm', fold=3)
# 2. ÏÉùÏÑ±Îêú Î™®Îç∏ÏùÑ tune_model()Ïóê Ï†ÑÎã¨ÌïòÏó¨ ÌäúÎãù
# n_iterÏôÄ fold Í∞íÏùÄ ÌïÑÏöîÏóê Îî∞Îùº Ï°∞Ï†àÌï† Ïàò ÏûàÏäµÎãàÎã§.
# n_iterÎäî RandomizedSearchCVÏùò ÌÉêÏÉâ ÌöüÏàòÏûÖÎãàÎã§. (Í∏∞Î≥∏Í∞íÏùÄ 10)
tuned_lightgbm = tune_model(lightgbm_model, n_iter=50, fold=2)
```

```python
# ÌäúÎãùÎêú Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
print(tuned_lightgbm)
```

```python
evaluate_model(tuned_lightgbm)
```

```python
# ÌäúÎãùÎêú Î™®Îç∏Ïùò Confusion Matrix (ÌòºÎèô ÌñâÎ†¨) ÏãúÍ∞ÅÌôî
plot_model(tuned_lightgbm, plot = 'confusion_matrix')
```

```python
# ÌäúÎãùÎêú Î™®Îç∏Ïùò AUC-ROC Í≥°ÏÑ† ÏãúÍ∞ÅÌôî
plot_model(tuned_lightgbm, plot = 'auc')
```

```python
# plot model
plot_model(tuned_lightgbm, plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```

```python
# Î™®Îç∏ ÏµúÏ¢Ö ÌïôÏäµ (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïö©)
final_model = finalize_model(tuned_lightgbm)

# ÏµúÏ¢Ö Î™®Îç∏ ÌôïÏù∏
print(final_model)
```

```python
# ÌïôÏäµÏóê ÏÇ¨Ïö©ÎêòÏßÄ ÏïäÏùÄ ÎØ∏ÏßÄÏùò Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏòàÏ∏°
# ÏòàÏãú: PyCaretÏù¥ ÎÇ¥Î∂ÄÏ†ÅÏúºÎ°ú Î∂ÑÎ¶¨Ìïú 20%Ïùò ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ÏÖãÏóê ÎåÄÌï¥ ÏòàÏ∏°
unseen_predictions = predict_model(final_model, data = s.data_unseen)

# ÏòàÏ∏° Í≤∞Í≥º ÌôïÏù∏
print(unseen_predictions)

# ÎßåÏïΩ ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ(new_data)Ïù¥ ÏûàÎã§Î©¥ ÏïÑÎûòÏôÄ Í∞ôÏù¥ ÏòàÏ∏°Ìï† Ïàò ÏûàÏäµÎãàÎã§.
# new_data = pd.DataFrame(...)
# new_predictions = predict_model(final_model, data = new_data)
```
