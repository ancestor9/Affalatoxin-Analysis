```python
import sys
import pycaret

# Python 버전이 3.11로 시작하는 것을 확인
print("파이썬 버전:", sys.version)

# PyCaret 버전 확인
print("PyCaret 버전:", pycaret.__version__)
```

```python
# Save the modified DataFrame to a pickle file
import pandas as pd
data= pd.read_csv('/teamspace/studios/this_studio/data.csv')
display(data.head())
print(data.shape)
```

```python
# fix_imbalance = True 파라미터를 추가하여 SMOTE 적용
# Import the classification module
from pycaret.classification import *
s = setup(data, 
          target = 'JDGMNT_WORD_NAME_encoded', 
          session_id = 123,
          fix_imbalance = True,
           rare_to_value = 0.1) # 전체 데이터에서 등장 비율이 10% 미만인 희귀 범주들을 하나로 묶어 새로운 범주로 변환)
```

```python
# 모든 모델을 비교하여 가장 성능이 좋은 모델을 찾습니다.
best_model = compare_models(fold=5)  # K-fold 5번만, default는 10번임

# best_model 변수에 가장 좋은 성능을 보인 모델이 저장됩니다.
print(best_model)
```

```python
# 1. 원하는 특정 모델 생성
# 모델 ID는 PyCaret의 모든 지원 모델 리스트에서 확인할 수 있습니다.
# LightGBM 모델을 3-겹 교차 검증으로 생성합니다.
lightgbm_model = create_model('lightgbm', fold=3)
# 2. 생성된 모델을 tune_model()에 전달하여 튜닝
# n_iter와 fold 값은 필요에 따라 조절할 수 있습니다.
# n_iter는 RandomizedSearchCV의 탐색 횟수입니다. (기본값은 10)
tuned_lightgbm = tune_model(lightgbm_model, n_iter=50, fold=2)
```

```python
# 튜닝된 모델의 성능을 확인할 수 있습니다.
print(tuned_lightgbm)
```

```python
evaluate_model(tuned_lightgbm)
```

```python
# 튜닝된 모델의 Confusion Matrix (혼동 행렬) 시각화
plot_model(tuned_lightgbm, plot = 'confusion_matrix')
```

```python
# 튜닝된 모델의 AUC-ROC 곡선 시각화
plot_model(tuned_lightgbm, plot = 'auc')
```

```python
# plot model
plot_model(tuned_lightgbm, plot = 'confusion_matrix', plot_kwargs = {'percent' : True})
```

```python
# 모델 최종 학습 (전체 데이터 사용)
final_model = finalize_model(tuned_lightgbm)

# 최종 모델 확인
print(final_model)
```

```python
# 학습에 사용되지 않은 미지의 데이터에 대한 예측
# 예시: PyCaret이 내부적으로 분리한 20%의 테스트 데이터셋에 대해 예측
unseen_predictions = predict_model(final_model, data = s.data_unseen)

# 예측 결과 확인
print(unseen_predictions)

# 만약 새로운 데이터프레임(new_data)이 있다면 아래와 같이 예측할 수 있습니다.
# new_data = pd.DataFrame(...)
# new_predictions = predict_model(final_model, data = new_data)
```
